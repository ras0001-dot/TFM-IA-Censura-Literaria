```python
# =========================
# 1. Instalar spaCy (solo la primera vez)
# =========================
!pip install spacy

# =========================
# 2. Importar librerรญas
# =========================
import re
import pandas as pd
from collections import Counter
from spacy.lang.es.stop_words import STOP_WORDS
```

```python

# =========================
# 3. Cargar texto
# =========================
text = open("Libro 1_norm.txt", encoding="utf-8").read()

# =========================
# 4. Normalizaciรณn
# =========================
text = text.lower()

# Quitar puntuaciรณn
text = re.sub(r"[^\w\sรกรฉรญรณรบรผรฑ]", " ", text)

# Quitar nรบmeros
text = re.sub(r"\d+", " ", text)

# =========================
# 5. Tokenizaciรณn
# =========================
words = text.split()

# =========================
# 6. Stopwords
# =========================
stopwords_extra = {
    "si","sรญ","ya","tan","solo","sรณlo","aun","aรบn","muy",
    "porque","pues","entonces","aquรญ","allรญ","tambiรฉn","pero",
    "dijo","dice","hacer","hace","asรญ","ser","estar",
    "cap"  # importante en tu caso
}

stopwords = STOP_WORDS.union(stopwords_extra)

# Filtrar palabras
words = [w for w in words if w not in stopwords and len(w) > 2]

# =========================
# 7. Crear aristas (window = 5)
# =========================
window = 5
pairs = []

for i in range(len(words)):
    for j in range(i+1, min(i+window, len(words))):
        pairs.append((words[i], words[j]))

# =========================
# 8. Contar pesos (frecuencia de cada par)
# =========================
pair_counts = Counter(pairs)

# =========================
# 9. Crear DataFrame con pesos
# =========================
df = pd.DataFrame([
    (source, target, weight)
    for (source, target), weight in pair_counts.items()
], columns=["Source", "Target", "Weight"])

# =========================
# 10. Guardar CSV
# =========================
output_file = "edges_libro_completo_window5_weighted.csv"
df.to_csv(output_file, index=False)

# =========================
# 11. Descargar
# =========================
from google.colab import files
files.download(output_file)
```
