```python
# Instalar spaCy (solo la primera vez)
!pip install spacy

import re
import pandas as pd
from spacy.lang.es.stop_words import STOP_WORDS

# -------------------------
# 1. Cargar texto
# -------------------------
text = open("libro 1_sam_20p.txt", encoding="utf-8").read()

# -------------------------
# 2. Normalización
# -------------------------
text = text.lower()

# Quitar puntuación y números
text = re.sub(r"[^\w\sáéíóúüñ]", " ", text)
text = re.sub(r"\d+", " ", text)

# -------------------------
# 3. Tokenización
# -------------------------
words = text.split()

# -------------------------
# 4. Stopwords
# -------------------------
stopwords_extra = {
    "si","sí","ya","tan","solo","sólo","aun","aún","muy",
    "porque","pues","entonces","aquí","allí","también","pero",
    "dijo","dice","hacer","hace","así","ser","estar"
}

stopwords = STOP_WORDS.union(stopwords_extra)

# Filtrar palabras
words = [w for w in words if w not in stopwords and len(w) > 2]

# -------------------------
# 5. Crear aristas (ventana = 5)
# -------------------------
window = 5
pairs = []

for i in range(len(words)):
    for j in range(i+1, min(i+window, len(words))):
        pairs.append((words[i], words[j]))

# -------------------------
# 6. Crear CSV
# -------------------------
df = pd.DataFrame(pairs, columns=["Source", "Target"])

df.to_csv("edges 1_window5.csv", index=False)

# -------------------------
# 7. Descargar
# -------------------------
from google.colab import files
files.download("edges 1_window5.csv")
```
