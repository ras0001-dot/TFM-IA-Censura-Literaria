```python
import re
import shutil
from pathlib import Path

base_norm = Path("/content/drive/MyDrive/Colab Notebooks/TFM/Normalizados paraextraer")
base_out = Path("/content/drive/MyDrive/Colab Notebooks/TFM/Capítulos")

# CONFIGURACIÓN
# metodo:
# "romanos"  -> capítulos I, II, III...
# "capitulo" -> líneas tipo CAPÍTULO I / Capítulo I. / CAPÍTULO PRIMERO
# "partes_romanos" -> Primera parte + capítulos I, II...
# "bloques" -> bloques fijos de texto
config = {
    7:  {"metodo": "bloques", "n_bloques": 10},
    17: {"metodo": "bloques", "n_bloques": 10},

    8:  {"metodo": "bloques", "n_bloques": 10},   # cambiar si luego ves capítulos
    18: {"metodo": "bloques", "n_bloques": 10},

    9:  {"metodo": "bloques", "n_bloques": 10},   # cambiar si luego ves capítulos
    19: {"metodo": "bloques", "n_bloques": 10},

    10: {"metodo": "bloques", "n_bloques": 10},   # cambiar si luego ves capítulos
    20: {"metodo": "bloques", "n_bloques": 10},

    16: {"metodo": "bloques", "n_bloques": 22},   # para hacerlo comparable con bloques_libro6 si 6 tiene 22 bloques
}

def limpiar_texto(text):
    return text.replace("\ufeff", "").strip()

def guardar_zip(output_dir):
    zip_base = str(output_dir)
    shutil.make_archive(zip_base, "zip", output_dir)
    return zip_base + ".zip"

def dividir_bloques(text, output_dir, n_bloques=10):
    output_dir.mkdir(parents=True, exist_ok=True)

    block_size = len(text) // n_bloques

    for i in range(n_bloques):
        start = i * block_size
        end = (i + 1) * block_size if i < n_bloques - 1 else len(text)

        block_text = text[start:end].strip()

        with open(output_dir / f"bloque_{i+1:02d}.txt", "w", encoding="utf-8") as f:
            f.write(block_text)

    return n_bloques

def dividir_por_patron(text, output_dir, pattern):
    output_dir.mkdir(parents=True, exist_ok=True)

    matches = list(re.finditer(pattern, text))

    if not matches:
        return 0

    for i, match in enumerate(matches):
        start = match.start()
        end = matches[i+1].start() if i+1 < len(matches) else len(text)

        chapter_text = text[start:end].strip()

        with open(output_dir / f"{i+1:03d}.txt", "w", encoding="utf-8") as f:
            f.write(chapter_text)

    return len(matches)

def dividir_partes_romanos(text, output_dir):
    output_dir.mkdir(parents=True, exist_ok=True)

    pattern_partes = r'(?im)^(primera|segunda|tercera|cuarta|quinta)\s+parte.*$'
    pattern_caps = r'(?m)^(?P<title>[IVXLCDM]+)\s*$'

    partes = list(re.finditer(pattern_partes, text))

    if not partes:
        return 0

    total_caps = 0

    for p_idx, parte in enumerate(partes):
        start_parte = parte.start()
        end_parte = partes[p_idx + 1].start() if p_idx + 1 < len(partes) else len(text)

        texto_parte = text[start_parte:end_parte]
        nombre_parte = parte.group(0).strip().replace(" ", "_").lower()

        dir_parte = output_dir / nombre_parte
        dir_parte.mkdir(parents=True, exist_ok=True)

        caps = list(re.finditer(pattern_caps, texto_parte))

        for c_idx, cap in enumerate(caps):
            start_cap = cap.start()
            end_cap = caps[c_idx + 1].start() if c_idx + 1 < len(caps) else len(texto_parte)

            texto_cap = texto_parte[start_cap:end_cap].strip()

            with open(dir_parte / f"{c_idx+1:03d}.txt", "w", encoding="utf-8") as f:
                f.write(texto_cap)

        total_caps += len(caps)

    return total_caps

def procesar_libro(num, cfg):
    input_file = base_norm / f"Libro {num}_norm.txt"
    metodo = cfg["metodo"]

    if metodo == "bloques":
        output_dir = base_out / f"bloques_libro{num}"
    elif metodo == "partes_romanos":
        output_dir = base_out / f"capitulos_libro{num}_partes"
    else:
        output_dir = base_out / f"capitulos_libro{num}"

    with open(input_file, "r", encoding="utf-8") as f:
        text = limpiar_texto(f.read())

    if metodo == "bloques":
        unidades = dividir_bloques(text, output_dir, cfg.get("n_bloques", 10))

    elif metodo == "romanos":
        pattern = r'(?m)^(?P<title>[IVXLCDM]+)\s*$'
        unidades = dividir_por_patron(text, output_dir, pattern)

    elif metodo == "capitulo":
        pattern = r'(?im)^(cap[ií]tulo\s+.+)$'
        unidades = dividir_por_patron(text, output_dir, pattern)

    elif metodo == "partes_romanos":
        unidades = dividir_partes_romanos(text, output_dir)

    else:
        raise ValueError(f"Método no reconocido: {metodo}")

    zip_path = guardar_zip(output_dir)

    print(f"Libro {num}: método={metodo}, unidades={unidades}, salida={output_dir.name}, zip={zip_path}")

for libro, cfg in config.items():
    procesar_libro(libro, cfg)
```

```python
# =========================
# 1. Librerías
# =========================
import re
from pathlib import Path

# =========================
# 2. Archivo de entrada
# =========================
input_file = "Libro 1_norm.txt"  # CAMBIA AQUÍ

# =========================
# 3. Carpeta de salida
# =========================
output_dir = Path("capitulos")
output_dir.mkdir(exist_ok=True)

# =========================
# 4. Leer texto
# =========================
with open(input_file, "r", encoding="utf-8") as f:
    text = f.read()

# =========================
# 5. Patrón capítulos (romanos en línea sola)
# =========================
pattern = r'(?m)^(?P<title>[IVXLCDM]+)\s*$'

matches = list(re.finditer(pattern, text))

# =========================
# 6. Corte por capítulos
# =========================
if not matches:
    print("❌ No se detectaron capítulos")
else:
    for i, match in enumerate(matches):
        start = match.start()
        end = matches[i+1].start() if i+1 < len(matches) else len(text)

        chapter_text = text[start:end].strip()
        chapter_title = match.group("title")

        filename = output_dir / f"{i+1:02d}_{chapter_title}.txt"

        with open(filename, "w", encoding="utf-8") as f:
            f.write(chapter_text)

    print(f"✅ {len(matches)} capítulos creados")
```

```python
# =========================
# 1. Librerías
# =========================
import re
from pathlib import Path

# =========================
# 2. Archivo entrada
# =========================
input_file = "Libro 3_norm.txt"

# =========================
# 3. Carpeta salida
# =========================
base_dir = Path("libro3_partes_capitulos")
base_dir.mkdir(exist_ok=True)

# =========================
# 4. Leer texto
# =========================
with open(input_file, "r", encoding="utf-8") as f:
    text = f.read()

# =========================
# 5. Detectar PARTES
# =========================
# Detecta: "Primera parte", "Segunda parte", etc.
pattern_partes = r'(?im)^(primera|segunda|tercera|cuarta|quinta)\s+parte.*$'

partes = list(re.finditer(pattern_partes, text))

if not partes:
    print("❌ No se detectaron partes")
else:
    for p_idx, parte in enumerate(partes):

        start_parte = parte.start()
        end_parte = partes[p_idx + 1].start() if p_idx + 1 < len(partes) else len(text)

        texto_parte = text[start_parte:end_parte]

        nombre_parte = parte.group(0).strip().replace(" ", "_").lower()

        # Crear carpeta de la parte
        dir_parte = base_dir / nombre_parte
        dir_parte.mkdir(exist_ok=True)

        # =========================
        # 6. Detectar capítulos dentro de la parte
        # =========================
        pattern_caps = r'(?m)^(?P<title>[IVXLCDM]+)\s*$'
        caps = list(re.finditer(pattern_caps, texto_parte))

        if not caps:
            print(f"⚠️ No capítulos en {nombre_parte}")
            continue

        for c_idx, cap in enumerate(caps):

            start_cap = cap.start()
            end_cap = caps[c_idx + 1].start() if c_idx + 1 < len(caps) else len(texto_parte)

            texto_cap = texto_parte[start_cap:end_cap].strip()
            cap_title = cap.group("title")

            filename = dir_parte / f"{c_idx+1:02d}_{cap_title}.txt"

            with open(filename, "w", encoding="utf-8") as f:
                f.write(texto_cap)

    print("✅ Partes y capítulos creados correctamente")
```

```python
# =========================
# 1. Librerías
# =========================
import re
from pathlib import Path

# =========================
# 2. Archivo de entrada
# =========================
input_file = "Libro 6_norm.txt"

# =========================
# 3. Carpeta de salida
# =========================
output_dir = Path("capitulos_libro6")
output_dir.mkdir(exist_ok=True)

# =========================
# 4. Leer texto
# =========================
with open(input_file, "r", encoding="utf-8") as f:
    text = f.read()

# =========================
# 5. Patrón FLEXIBLE
# =========================
pattern = r'(?im)^\s*(cap(?:\.|[íi]tulo)?\s*\.?\s*(?:\d+|[IVXLCDM]+).*)$'

matches = list(re.finditer(pattern, text))

print(f"🔎 Capítulos detectados: {len(matches)}")

# =========================
# 6. Crear capítulos
# =========================
if not matches:
    print("❌ No se detectaron capítulos")
else:
    for i, match in enumerate(matches):
        start = match.start()
        end = matches[i+1].start() if i+1 < len(matches) else len(text)

        chapter_text = text[start:end].strip()
        chapter_title = match.group(1)

        # nombre corto para evitar error de longitud
        filename = output_dir / f"{i+1:03d}.txt"

        with open(filename, "w", encoding="utf-8") as f:
            # guardamos título dentro del archivo
            f.write(chapter_title + "\n\n" + chapter_text)

    print(f"✅ {len(matches)} capítulos creados")
```

```python
from pathlib import Path

# Carpeta con capítulos ya separados
input_dir = Path("capitulos_libro6")

# Carpeta de salida
output_dir = Path("bloques_libro6")
output_dir.mkdir(exist_ok=True)

# Cambia a 12 si prefieres bloques más grandes
block_size = 12

# Ordenar capítulos
chapter_files = sorted(input_dir.glob("*.txt"))

print(f"Capítulos encontrados: {len(chapter_files)}")

for i in range(0, len(chapter_files), block_size):
    block_files = chapter_files[i:i + block_size]

    block_text = ""

    for file in block_files:
        with open(file, "r", encoding="utf-8") as f:
            block_text += f.read().strip() + "\n\n"

    block_number = i // block_size + 1
    output_file = output_dir / f"bloque_{block_number:02d}.txt"

    with open(output_file, "w", encoding="utf-8") as f:
        f.write(block_text.strip())

print(f"Bloques creados: {len(list(output_dir.glob('*.txt')))}")
```
